# STEP- 3: Data Preparation for Model Building

**Removing unwanted features form the dataset based on following theory & assumptions:**

1) "casual" & "registered" : Both these columns contains the count of bike booked by different categories of customers. Since our objective is to find the total count of bikes i.e. "cnt" and not by specific category, we will ignore these two columns.

2) "instant" : Its only an index value.

3) "dteday" : The month & year features are already given in the dataset. Hence this feature is redundent. Therefore it will be dropped

```python
bikeshare_df.columns
```

# STEP- 4: Splitting the dataset

```python
# Importing libraries for model building

import sklearn
from sklearn.model_selection import train_test_split
```

```python
#Splitting the Data into Training and Testing Sets

data_train, data_test = train_test_split(bikeshare_df_new, train_size =0.70, test_size = 0.30, random_state = 333)
data_train
```

```python
#Applying scaler to all numerical features

var_scale =['temp','atemp','hum','windspeed','cnt']
data_train[var_scale] =scaler.fit_transform(data_train[var_scale]) 
data_train.head()
```

# STEP- 5: Model Building

```python
# taking out the traget variable from the training dataset

y_train = data_train.pop('cnt')
X_train = data_train
```

```python
# Importing RFE(Recursive Feature Elimination) & linear RRegeression from scikit learn

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Fitting the model 

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Running RFE with the output number of the variable equal to 15

rfe = RFE(estimator=lm, n_features_to_select=18)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Checking the variables whoch are selected by rfe & storing it
col = X_train.columns[rfe.support_]
col
```

```python
# Checking the variables whoch are not selected by rfe

X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

### Model 1

```python
# Check for the VIF values of the feature variables 

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# using stats model

import statsmodels.api as sm

# Adding a constant
X_train_lm1 = sm.add_constant(X_train_rfe)

# Creating first fitted model
lr1 = sm.OLS(y_train, X_train_lm1).fit()
```

```python
# Checking the parameters obtained

lr1.params
```

```python
# Printing  a summary of the linear regression model obtained

print(lr1.summary())
```

### Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), plotting the histogram of the error terms and to check what it looks like.

```python
y_train_pred = lr9.predict(X_train_lm9)
```

```python
# Plot the histogram of the error terms

fig = plt.figure()
sns.distplot(y_train-y_train_pred, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
plt.show()
```

From the above histogram, we could see that the **Residuals** are **normally distributed**. Hence our assumption for Linear Regression is valid.

### Checking for patterns in residuals

```python
plt.scatter(y_train_pred,(y_train-y_train_pred))
plt.xlabel('Predicted Values', fontsize = 18) 
plt.ylabel('Errors', fontsize = 18) 
plt.show()
```

### In the above graph it is clear that residuals are independent. Hence no relation between predictied values & residuals.
